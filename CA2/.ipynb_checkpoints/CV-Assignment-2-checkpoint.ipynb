```python
def rgb2hsi(na):
    with np.errstate(divide='ignore', invalid='ignore'):
        #na = np.array(im,dtype='float32')
        na = na/255
        r,g,b = na[:,:,0], na[:,:,1], na[:,:,2]
        I = (r+g+b)/3
        S = 1 - np.minimum(np.minimum(r,g),b)/I
        H = np.zeros(np.shape(r))
        rc = np.shape(r)
        for i in range(rc[0]):
            for j in range(rc[1]):
                if not (r[i,j]==g[i,j] and g[i,j]==b[i,j]):
                    #print("r,g,b",r[i,j], g[i,j], b[i,j])
                    num = 0.5*(r[i,j]-g[i,j]+r[i,j]-b[i,j])
                    den = (r[i,j]-g[i,j])**2 + (r[i,j]-g[i,j])*(g[i,j]-b[i,j])
                    den = np.clip(den,1e-5,1)
                    den = np.sqrt(den)
                    #print("num,den",num,den)
                    x = num/den
                    #print("x",x)
                    theta = math.acos(np.clip(x,1e-5,1))
                    #theta = np.rad2deg(theta)
                    #print("theta",theta)
                    if b[i,j] > g[i,j]:
                        theta = 2*math.pi - theta

                    H[i,j] = np.rad2deg(theta)
                else:
                    H[i,j] = 0
        
        S[np.isnan(S)] = 0
                    
#         for i in range(0, rc[0]):
#             for j in range(0, rc[1]):
#                 theta = 0.5 * ((r[i][j] - g[i][j]) + (r[i][j] - b[i][j])) / \
#                             math.sqrt((r[i][j] - g[i][j])**2 +
#                                     ((r[i][j] - b[i][j]) * (g[i][j] - b[i][j])))
#                 H[i][j] = math.acos(theta)

#                 if b[i][j] <= g[i][j]:
#                     H[i][j] = H[i][j]
#                 else:
#                     H[i][j] = ((360 * math.pi) / 180.0) - H[i][j]
                    
        return [H,S,I]
```

```python
def RGB_TO_HSI(img):
    with np.errstate(divide='ignore', invalid='ignore'):

        #Load image with 32 bit floats as variable type
        bgr = np.float32(img)/255

        #Separate color channels
        blue = bgr[:,:,0]
        green = bgr[:,:,1]
        red = bgr[:,:,2]

        #Calculate Intensity
        
        intensity = np.divide(blue + green + red, 3)
        #Calculate Saturation
        
        minimum = np.minimum(np.minimum(red, green), blue)
        saturation = 1 - (3 / (red + green + blue + 0.001) * minimum)


        #Calculate Hue
        hue = np.copy(red)

        for i in range(0, blue.shape[0]):
            for j in range(0, blue.shape[1]):
                hue[i][j] = 0.5 * ((red[i][j] - green[i][j]) + (red[i][j] - blue[i][j])) / \
                            math.sqrt((red[i][j] - green[i][j])**2 +
                                    ((red[i][j] - blue[i][j]) * (green[i][j] - blue[i][j])))
                hue[i][j] = math.acos(hue[i][j])

                if blue[i][j] <= green[i][j]:
                    hue[i][j] = hue[i][j]
                else:
                    hue[i][j] = ((360 * math.pi) / 180.0) - hue[i][j]


        #Merge channels into picture and return image
        #hsi = cv2.merge((calc_hue(red, blue, green), calc_saturation(red, blue, green), calc_intensity(red, blue, green)))
        return [hue,saturation,intensity]
```

```python
def get_avg_hsi(l):
    #print(list(map(lambda x : round(x,2), list(map(np.mean, l)))))
    return list(map(lambda x : round(x,2), list(map(np.mean, l))))
```

## Task 1
Differences of average hue, saturation and indensities between I1 and I2 for all ten images.

I1 = I(x,y)

I2 = I(x,y) * p where p in [0,1]


```python
for i in range(1,10+1):
    img_name = f"{i}.tiff"
    I = Image.open("10images/"+img_name)
    I = np.array(I,dtype='float32')
    I1 = I
    I2 = I*0.5
    h1,s1,i1 = get_avg_hsi(rgb2hsi(I1))
    h2,s2,i2 = get_avg_hsi(rgb2hsi(I2))
    print(f"Image : {img_name}")
    print(f"Avg H of I1 : {h1:.2f}, Avg S of I1: {s1:.2f}, Avg I of I1 : {i1:.2f}")
    print(f"Avg H of I2 : {h2:.2f}, Avg S of I2: {s2:.2f}, Avg I of I2: {i2:.2f}")
    print(f"Diffeence in Hue : {(abs(h1-h2)):.2f}, Difference in Saturation : {(abs(s1-s2)):.2f}, Difference in Intensity : {(abs(i1-i2)):.2f}")
    print("-----------------------------------------------")
```

## Task2
Differences of average hue, saturation and indensities between I1 and I2 for all ten images.

I1 = I(x,y)

I2 = I(x,y) + [G(x,y)-R(x,y), B(x,y) - G(x,y), R(x,y) - B(x,y)]

```python
for i in range(1,10+1):
    img_name = f"{i}.tiff"
    I = Image.open("10images/" + img_name)
    I = np.array(I, dtype='float32')
    I1 = I
    I2 = np.zeros(np.shape(I))
    I2[:,:,0] += I[:,:,1]-I[:,:,0]
    I2[:,:,1] += I[:,:,2]-I[:,:,1]
    I2[:,:,2] += I[:,:,0]-I[:,:,2]
    I2 = np.clip(I2,0,255)
    h1,s1,i1 = get_avg_hsi(rgb2hsi(I1))
    h2,s2,i2 = get_avg_hsi(rgb2hsi(I2))
    print(f"Image : {img_name}")
    print(f"Avg Hue I1 : {h1:.2f}, Avg Saturation I1: {s1:.2f}, Avg Intensity I1 : {i1:.2f}")
    print(f"Avg Hue I2 : {h2:.2f}, Avg Saturation I2: {s2:.2f}, Avg Intensity I2 : {i2:.2f}")
    print(f"Diffeence in Hue : {(abs(h1-h2)):.2f}, Difference in Saturation : {(abs(s1-s2)):.2f}, Difference in Intensity : {(abs(i1-i2)):.2f}")
    print("------------------------------------------------------------------")
```

## Task3
Differences of average hue, saturation and indensities between I1 and I2 for all ten images.

I1 = I(x,y)

I2 = I'(x,y) where I' is AWB of I1

```python
for i in range(1,10+1):
    img_name = f"{i}.tiff"
    I = Image.open("10images/" + img_name)
    I = np.array(I, dtype='float32')
    I1 = I.copy()
    I2 = np.zeros(np.shape(I))
    avg_r = np.mean(I1[:,:,0])
    avg_g = np.mean(I1[:,:,1])
    avg_b = np.mean(I1[:,:,2])
    m = np.mean([avg_r, avg_g, avg_b])
    I2[:,:,0] = I1[:,:,0]*(m/avg_r)
    I2[:,:,1] = I1[:,:,1]*(m/avg_g)
    I2[:,:,2] = I1[:,:,2]*(m/avg_b)
    I2 = np.clip(I2,0,255)
    h1,s1,i1 = get_avg_hsi(rgb2hsi(I1))
    h2,s2,i2 = get_avg_hsi(rgb2hsi(I2))
    print(f"Image : {img_name}")
    print(f"Avg Hue I1 : {h1:.2f}, Avg Saturation I1: {s1:.2f}, Avg Intensity I1 : {i1:.2f}")
    print(f"Avg Hue I2 : {h2:.2f}, Avg Saturation I2: {s2:.2f}, Avg Intensity I2 : {i2:.2f}")
    print(f"Diffeence in Hue : {(abs(h1-h2)):.2f}, Difference in Saturation : {(abs(s1-s2)):.2f}, Difference in Intensity : {(abs(i1-i2)):.2f}")
    print("------------------------------------------------------------------")
```
